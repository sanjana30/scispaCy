**Matrix for Entities using the scispaCy bc5cdr model**

```
# Read the entities files for Bc5cdr model that was created earlier
entities_df = pd.read_csv('/content/en_ner_bc5cdr_md.csv')
print(entities_df.loc[0, 'ENTITY'])
print(entities_df.iloc[0])
```

```
# Create a list of all the entities combined and store it in a file for future reference
entitiesList=[]
for i in range(0, 1):
  templist = entities_df.loc[i, 'ENTITY'].strip('()').split(',')
  for j in range(0, len(templist)):
    entitiesList.append(templist[j])    
print(entitiesList)
print('Length of entities list: ', len(entitiesList))
# Clean this list to remove any duplicacy
entitiesList=list(dict.fromkeys(entitiesList))
print('Length of clean entities list: ', len(entitiesList))

entitiesList.append('headache')
entitiesList.append('oxygen')
print('Length of clean entities list: ', len(entitiesList))
```

```
# Create a dictionary that assigns a group entities derived from one note text a unique key
entitiesDict = {}
for i in range(0, 1):
  temp = entities_df.loc[i, 'ENTITY'].strip('()').split(',')
  entitiesDict[i]=temp
  
entitiesDict[1]=['oxygen', 'headache']
print(entitiesDict)
print('Length of entities dictionary is: ',len(entitiesDict))
```

```
# Create a basic zero vector matrix of dimensionality NxN where N represents the length of the clean entities list
print('Dimension of the matrix being built is: ', len(entitiesList),'x',len(entitiesList))
rows, cols = (len(entitiesList), len(entitiesList)) 
arr=[] 
for i in range(cols): 
    col = [] 
    for j in range(rows): 
        col.append(0) 
    arr.append(col) 

```

```
# Assign values to the matrix based on the appearance of entities in the same note text
flag=0

# entitiesList=['cat', 'dog', 'mammal']
N=len(entitiesList)
w, h = N, N;
arr = [[0 for x in range(w)] for y in range(h)] 


print(entitiesList)
for i in range(len(entitiesList)):
  for j in range(len(entitiesList)):
    if(entitiesList[j] == entitiesList[i]):
      arr[i][j]=1
    else:
      iKeys = []
      jKeys = []
      for key, value in entitiesDict.items():
        if(entitiesList[j] in value): 
          jKeys.append(key)
        if(entitiesList[i] in value): 
          iKeys.append(key)
      if((( set(iKeys).issubset(set(jKeys)) ) or ( set(jKeys).issubset(set(iKeys)) )) and (len(iKeys)!= 0 and len(jKeys)!=0) ): 
        arr[i][j]=1

print(arr[0])
print(arr[1])
print(arr[2])
print(arr[3])
print(arr[4])
print(arr[5])
```
